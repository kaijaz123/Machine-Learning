```python
import pandas as pd 
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import train_test_split
```

```python
# Make a simple DataFrame of the given data 
data = pd.DataFrame({"C1":[1,2,3,4,5,6], 
                    "C2":[4,5,6,7,8,9], 
                    "C3":[0,0,0,0,0,0],
                    "C4":[1,1,1,1,1,1]})
data.head()
```

```python
var_thres=VarianceThreshold(threshold=1)
var_thres.fit(data)
```

```python
data.columns[var_thres.get_support()]
# C1 and C2 is above the threshold of variance 1
```

```python
df = pd.read_csv('santander-customer-satisfaction/train.csv',nrows=1000)
```

```python
df.head()
# total 371 columns at first
```

```python
X = df.drop('TARGET',axis=1)
y = df['TARGET']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state=10)
x_train.shape,y_train.shape
```

```python
# set the threshold to 1 then fit it
var_thres=VarianceThreshold(threshold=1)
var_thres.fit(x_train)
```

```python
len(var_thres.get_support()[var_thres.get_support() == False])
# 243 columns that variance less than 1
```

```python
constant_columns = [column for column in x_train.columns 
                    if column not in x_train.columns[var_thres.get_support()]]
constant_columns
```

```python
new_x_train = x_train.drop(columns=constant_columns,axis=1)
```
